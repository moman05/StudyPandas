#### 하나의 열이 여러 의미를 가지는 경우
 - split()과 concat()으로 데이터 정리
 - 에볼라 바이러스 감염 및 사망자 수

```python
ebola = pd.read_csv('C:/STUDY/data/country_timeseries.csv')
print(ebola.columns)
```

```python
print(ebola.iloc[:5, [0,1,2,3,10,11]])
```

```python
# date, dat 열 고정 / 나머지 열 피벗
ebola_long = pd.melt(ebola, id_vars=['Date', 'Day'])
print(ebola_long.head())
```

```python
# 'variable' 열의 데이터를 _ 기준으로 분리
variable_split = ebola_long.variable.str.split('_')
print(variable_split[:5])
```

```python
# 분리된 데이터 확인
status_values = variable_split.str.get(0)
country_values = variable_split.str.get(1)

print(status_values[:5])
print(country_values[:5])
```

```python
# 1. 분리된 데이터를 'ebola_long' 데이터 프레임에 추가
ebola_long['status'] = status_values
ebola_long['country'] = country_values
print(ebola_long.head())
```

```python
# 2. 분리된 데이터를 concat()으로 합치기
variable_split = ebola_long.variable.str.split('_', expand=True)
variable_split.columns = ['status2', 'country2']
ebola_parsed = pd.concat([ebola_long, variable_split], axis=1)

print(ebola_parsed.head())
```

```python
ebola = ebola_parsed.loc[:, ['Date', 'Day', 'status', 'country', 'value']]
ebola
```

```python
# 빌보드 차트 데이터 가져오기
billboard = pd.read_csv('C:/STUDY/data/billboard.csv')
billboard.head()
```

```python
# 빌보드 차트 데이터 피벗
billboard_long = pd.melt(
    billboard,
    id_vars=['year','artist','track','time','date.entered'],
    var_name='week', value_name='rating'
)

print(billboard_long.shape)
billboard_long.head(10)
```

```python
# 노래제목으로 중복 데이터 현황 확인
billboard_long[billboard_long.track == 'Loser'].head()
```

#### 빌보드 차트 데이터의 중복 데이터 처리하기(정규화)

```python
# 'year', 'artist','track','time' 중복 데이터를 새로운 데이터프레임으로 저장
billboard_songs = billboard_long[['year','artist','track','time']]
print(billboard_songs.shape)
```

```python
# 중복 데이터 제거 (drop_duplicates())
billboard_songs = billboard_songs.drop_duplicates()
print(billboard_songs.shape)
```

```python
# 순번(id) 추가
billboard_songs['id'] = range(len(billboard_songs))
billboard_songs.head(10)
```

```python
# 노래 정보와 주간 순위 데이터 합치기
billboard_ratings = billboard_long.merge(
    billboard_songs, on=['year','artist','track','time']
)
print(billboard_ratings.shape)
billboard_ratings.head()
```
